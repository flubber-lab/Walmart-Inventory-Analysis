```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Walmart.csv")
df
```

```python
df.set_index('transaction_id', inplace=True)
df
```

```python
df.info()
```

```python
df['promotion_type'] = np.where(df['promotion_applied']== False, 'No Discount',np.where(
    df['promotion_type'].isna(), 'Unknown Discount', df['promotion_type']
))
df
```

```python
df.info()
```

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df.info()
```

```python
import plotly.express as px
```

```python
sns.histplot(df['unit_price'], bins = 50, kde=True)
```

```python
import plotly.io as pio
pie=px.pie(df, values='quantity_sold', names='product_name', title='Quantity Sold by Product',
       color_discrete_sequence=px.colors.qualitative.Plotly,hole=0.3)
pio.write_image(pie, "Quantity Sold by Product.png")
```

```python
x = px.bar(df,x='product_name',y='unit_price')
x.update_layout(xaxis_title='Product Name',yaxis_title='Unit Price',title='Unit Price by Product',legend_title='Category')
x.show()
pio.write_image(x, "Unit Price by Product.png")
```

```python
y=px.bar(df,x='product_name',color='payment_method', barmode='group',
       title='Payment Method by Product',color_discrete_sequence=px.colors.qualitative.Pastel)
y.update_layout(xaxis_title='Product Name',yaxis_title='Count',legend_title='Payment Method')
y.show()
pio.write_image(y, "Payment Method by Product.png")
```

```python
# Group by Product and PaymentMethod, and calculate total revenue
grouped_df = df.groupby(['product_name', 'payment_method'], as_index=False)['unit_price'].sum()

fig = px.bar(grouped_df, x='product_name', y='unit_price', color='payment_method', barmode='stack',
             title='Revenue by Product and Payment Method',color_discrete_sequence=px.colors.qualitative.Vivid)
fig.update_layout(xaxis_title='Product Name', yaxis_title='Revenue', legend_title='Payment Method')
fig.show()
pio.write_image(fig, "Revenue by Product and Payment Method.png")
```

```python
#store location and revenue

slr = df.groupby(['store_location'],as_index=False)['unit_price'].sum()
slr_chart=px.bar(slr,x='store_location',y='unit_price',title='Store Location vs Revenue',
      color_discrete_sequence=px.colors.qualitative.Pastel)
slr_chart.update_layout(xaxis_title='Store Location',yaxis_title='Revenue')
slr_chart.show()
pio.write_image(slr_chart, "Store Location vs Revenue.png")
```

```python
#Product ID vs Inventory Level
PIIL=px.bar(df,x='product_id',y='inventory_level',color='product_name',title='Product ID vs Inventory Level')
PIIL.update_layout(xaxis_title='Product ID',yaxis_title='Inventory Level')
PIIL.show()
pio.write_image(PIIL, "Product ID vs Inventory Level.png")
```

```python
df['transaction_date']= pd.to_datetime(df['transaction_date']).dt.date
df
```

```python
store = df.groupby(['store_location','product_name'], as_index=False)['quantity_sold'].sum()
chart = px.bar(store, x='store_location',y='quantity_sold', color='product_name',
               title='Product Sales by Store Location',color_discrete_sequence=px.colors.qualitative.Pastel)
chart.update_layout(xaxis_title='Store Location',yaxis_title='Quantity Sold', legend_title='Product Name')
chart.show()
pio.write_image(chart, "Product Sales by Store Location.png")
```
